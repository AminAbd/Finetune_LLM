```python
from datasets import load_dataset, DatasetDict, Dataset

from transformers import (
    AutoTokenizer,
    AutoConfig, 
    AutoModelForSequenceClassification,
    DataCollatorWithPadding,
    TrainingArguments,
    Trainer)

from peft import PeftModel, PeftConfig, get_peft_model, LoraConfig
import evaluate
import torch
import numpy as np
```

Model

```python
model_checkpoint = 'distilbert-base-uncased'
#model_checkpoint ='bert-base-uncased'
# model_checkpoint = 'roberta-base' # you can alternatively use roberta-base but this model is bigger thus training will take longer

# define label maps
id2label = {0: "Negative", 1: "Positive"}
label2id = {"Negative":0, "Positive":1}

# generate classification model from model_checkpoint
model = AutoModelForSequenceClassification.from_pretrained(
    model_checkpoint, num_labels=2, id2label=id2label, label2id=label2id)
```

```python
# load dataset
dataset = load_dataset('shawhin/imdb-truncated')
dataset
```

```python
# display % of training data with label=1
np.array(dataset['train']['label']).sum()/len(dataset['train']['label'])
```

```python
# display architecture
model
```

preprocess data

```python
# create tokenizer
# Use local_files_only=True to avoid chat template check (files should be cached from model loading)
from transformers import DistilBertTokenizer

tokenizer = DistilBertTokenizer.from_pretrained(
    model_checkpoint, 
    add_prefix_space=True, 
    local_files_only=True
)

# add pad token if none exists
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})
    model.resize_token_embeddings(len(tokenizer))
```

```python
import sys, transformers, huggingface_hub
print("PYTHON:", sys.executable)
print("transformers:", transformers.__version__)
print("huggingface_hub:", huggingface_hub.__version__)
print("model_checkpoint:", model_checkpoint)
```

```python
# create tokenize function
def tokenize_function(examples):
    # extract text
    text = examples["text"]

    #tokenize and truncate text
    tokenizer.truncation_side = "left"
    tokenized_inputs = tokenizer(
        text,
        return_tensors="np",
        truncation=True,
        max_length=512
    )

    return tokenized_inputs
```

```python
# tokenize training and validation datasets
tokenized_dataset = dataset.map(tokenize_function, batched=True)
tokenized_dataset
```

```python
# create data collator
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

evaluation

```python
# import accuracy evaluation metric
accuracy = evaluate.load("accuracy")
```

```python
# define an evaluation function to pass into trainer later
def compute_metrics(p):
    predictions, labels = p
    predictions = np.argmax(predictions, axis=1)

    return {"accuracy": accuracy.compute(predictions=predictions, references=labels)}
```

Apply untrained model to text

```python
# define list of examples
text_list = ["It was good.", "Not a fan, don't recommed.", "Better than the first one.", "This is not worth watching even once.", "This one is a pass."]

print("Untrained model predictions:")
print("----------------------------")
for text in text_list:
    # tokenize text
    inputs = tokenizer.encode(text, return_tensors="pt")
    # compute logits
    logits = model(inputs).logits
    # convert logits to label
    predictions = torch.argmax(logits)

    print(text + " - " + id2label[predictions.tolist()])
```

Train model

```python
peft_config = LoraConfig(task_type="SEQ_CLS",
                        r=4,
                        lora_alpha=32,
                        lora_dropout=0.01,
                        target_modules = ['q_lin'])
```

```python
peft_config
```

```python
model = get_peft_model(model, peft_config)
model.print_trainable_parameters()
```

```python
# hyperparameters
lr = 1e-3
batch_size = 4
num_epochs = 10
```

```python
# define training arguments
training_args = TrainingArguments(
    output_dir= model_checkpoint + "-lora-text-classification",
    learning_rate=lr,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    num_train_epochs=num_epochs,
    weight_decay=0.01,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
)
```

```python
# creater trainer object
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator, # this will dynamically pad examples in each batch to be equal length
    compute_metrics=compute_metrics,
)

# train model
trainer.train()
```

```python
# Save the trained model to directory
import os
save_directory = "./Finetuned_model"
os.makedirs(save_directory, exist_ok=True)
model.save_pretrained(save_directory)
tokenizer.save_pretrained(save_directory)
print(f"Model saved to {save_directory}")
```

```python
# Load the trained model
from transformers import DistilBertTokenizer, AutoModelForSequenceClassification
from peft import PeftModel

# Load base model
base_model = AutoModelForSequenceClassification.from_pretrained(
    model_checkpoint, 
    num_labels=2, 
    id2label=id2label, 
    label2id=label2id
)

# Load the PEFT/LoRA adapter
loaded_model = PeftModel.from_pretrained(base_model, "./Finetuned_model")

# Load tokenizer
loaded_tokenizer = DistilBertTokenizer.from_pretrained("./Finetuned_model")

print("Model loaded successfully!")
loaded_model.eval()  # Set to evaluation mode
```

Generate prediction

```python
# Detect device (use 'cpu' for Windows, 'cuda' if GPU available, 'mps' for Mac)
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

print("Trained model predictions:")
print("--------------------------")
for text in text_list:
    inputs = tokenizer.encode(text, return_tensors="pt").to(device)

    logits = model(inputs).logits
    predictions = torch.max(logits,1).indices

    print(text + " - " + id2label[predictions.tolist()[0]])
```
